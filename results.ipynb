```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# RQ1

# baseline across different training data sizes
# baseline = [0.56, 0.68, 0.88, 0.9]
baseline = [0.56, 0.68, 0.88]

uda = [0.75, 0.85, 0.88]
eda = [0.57, 0.73, 0.88]
ssmba = [0.62, 0.7, 0.87]
rotom = [0.6, 0.82, 0.88]
latent = [0.53, 0.78, 0.89]
spelling = [0.56, 0.74, 0.88]
chat = [0.67, 0.8, None]


x = [20, 100, 1000]
plt.xticks(x, rotation=45)
plt.plot(x, baseline, '-o', label='Baseline')
plt.plot(x, uda, '-o', label='UDA')
plt.plot(x, eda, '-o', label='EDA')
plt.plot(x, ssmba, '-o', label='SSMBA')
plt.plot(x, rotom, '-o', label='Rotom')
plt.plot(x, latent, '-o', label='Latent')
plt.plot(x, spelling, '-o', label='Spelling')
plt.plot(x, chat, '-o', label='ChatGPT')
plt.legend()
```

```python
def compute_improve(base, new):
    base = np.array(base)
    new = np.array(new)
    return (new - base) / base

uda_improve = compute_improve(baseline, uda)
eda_improve = compute_improve(baseline, eda)
ssmba_improve = compute_improve(baseline, ssmba)
rotom_improve = compute_improve(baseline, rotom)
latent_improve = compute_improve(baseline, latent)
spelling_improve = compute_improve(baseline, spelling)
# chat_improve = compute_improve(baseline, chat)
chat_improve = compute_improve(baseline[:-1], chat[:-1])

# x = [20, 100, 1000]
x = [0, 1, 2]
# plt.xticks(x, rotation=45)
# plt.plot(x, uda_improve, '-o', label='UDA Improve')
# plt.plot(x, eda_improve, '-o', label='EDA Improve')
# plt.plot(x, ssmba_improve, '-o', label='SSMBA Improve')
# plt.plot(x, rotom_improve, '-o', label='Rotom Improve')
# plt.plot(x, latent_improve, '-o', label='Latent Improve')
# plt.plot(x, spelling_improve, '-o', label='Spelling Improve')
# plt.plot(x[:-1], chat_improve, '-o', label='ChatGPT Improve')
# plt.legend()


fig, (ax1, ax2, ax3, ax4, ax5, ax6, ax7) = plt.subplots(1, 7, figsize=(12, 2), sharey=True)

# ax1.plot(x, uda_improve, '-o', label='UDA Improve')
# ax2.plot(x, eda_improve, '-o', label='EDA Improve')
# ax3.plot(x, ssmba_improve, '-o', label='SSMBA Improve')
# ax4.plot(x, rotom_improve, '-o', label='Rotom Improve')
# ax5.plot(x, latent_improve, '-o', label='Latent Improve')
# ax6.plot(x, spelling_improve, '-o', label='Spelling Improve')
# ax7.plot(x[:-1], chat_improve, '-o', label='ChatGPT Improve')

colors = ['#0059b3', '#ff7733', '#33cc33']
ax1.bar(x, uda_improve, color=colors, label='UDA Improve')
ax2.bar(x[:-1], chat_improve, color=colors, label='ChatGPT Improve')

ax3.bar(x, rotom_improve, color=colors, label='Rotom Improve')
ax4.bar(x, latent_improve, color=colors, label='Latent Improve')
ax5.bar(x, ssmba_improve, color=colors, label='SSMBA Improve')

ax6.bar(x, eda_improve, color=colors, label='EDA Improve')
ax7.bar(x, spelling_improve, color=colors, label='Spelling Improve')

ax1.set_xticks([0,1,2])
ax2.set_xticks([0,1,2])
ax3.set_xticks([0,1,2])
ax4.set_xticks([0,1,2])
ax5.set_xticks([0,1,2])
ax6.set_xticks([0,1,2])
ax7.set_xticks([0,1,2])

ax1.set_xticklabels(['S', 'M', 'L'])
ax2.set_xticklabels(['S', 'M', 'L'])
ax3.set_xticklabels(['S', 'M', 'L'])
ax4.set_xticklabels(['S', 'M', 'L'])
ax5.set_xticklabels(['S', 'M', 'L'])
ax6.set_xticklabels(['S', 'M', 'L'])
ax7.set_xticklabels(['S', 'M', 'L'])

ax1.set_title('UDA')
ax2.set_title('ChatGPT')
ax3.set_title('Rotom')
ax4.set_title('Latent')
ax5.set_title('SSMBA')
ax6.set_title('EDA')
ax7.set_title('Spelling')


# ax1.set_xlabel('Dataset Size')
# ax1.set_ylabel('Improvement')
# ax1.set_title('Model Performance Improvement with Data Augmentation (Left)')
# ax2.set_xlabel('Dataset Size')
# ax2.set_ylabel('Improvement')
# ax2.set_title('Model Performance Improvement with Data Augmentation (Right)')


# Set unified title
fig.suptitle("Comparison of Improvements in F1 for Different Models on Different Dataset Sizes", fontsize=14)

# Set x-axis label
fig.text(0.5, -0.1, "Dataset Size", ha='center', fontsize=14)


# plt.legend()
plt.tight_layout()
plt.show()
```
